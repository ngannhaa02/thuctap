```python
conn = sqlite3.connect('ax.db')
query2 = 'SELECT * FROM dataSales'
df2 = pd.read_sql_query(query2, conn)
conn.close()
```

- bảng này mô tả dữ liệu bán hàng gồm 4248 dòng và 20 cột
- có 3 kiểu dữ liệu : float64, int64, object
- các thuộc tính trong bảng dữ liệu không bị thiếu

```python
df2.isnull().sum() #đếm giá trị thiếu trong từng cột -> không có giá trị thiếu 
```

```python
df2.duplicated().sum() # không có giá trị trùng lặp
```

```python
# kiểm tra giá trị cột 'Profit' =  sale (tổng doanh số) - COGS (giá vốn bán) - Total Expense (tổng chi phí)
for i in range(len(df2)):
   if( df2['Sales'][i]- df2['COGS'][i] - df2['Total Expenses'][i] != df2["Profit"][i]):
      print("test")
```

```python
# cập nhật lại Profit ở bảng dữ liệu df2
df2['Profit'] = df2['Sales'] - df2['COGS'] - df2['Total Expenses']
```

```python
# kiểm tra cột 'Margin' =  cột 'Profit' / cột 'Sales' không
for i in range(len(df2)):
   if( df2["Profit"][i] / df2["Sales"][i]!=df2["Margin"][i]):
      print("test")
```

```python
# cập nhật lại cột Margin 
df2["Margin"] = df2["Profit"] / df2["Sales"] 
```

```python
# kiểm tra xem tổng lợi nhuận (Profit) và tổng chi phí (Total Expenses) = giá trị biên ngân sách (Budget Margin) hay không
for i in range(len(df2)):
   if( df2["Profit"][i]+df2["Total Expenses"][i]!=df2["Budget Margin"][i]):
      print("test")
```

```python
# cập nhật lại cột Budget Margin
df2["Budget Margin"] = df2["Profit"] + df2["Total Expenses"]
```

```python
# Chuyển đổi cột 'Date' thành định dạng ngày tháng hợp lệ
df2['Date'] = pd.to_datetime(df2['Date'], errors='coerce')
```
